```python
for feature in data.columns :
    
    dataset=data.copy()
    if feature!='quality':
        plt.scatter(dataset[feature],dataset['quality'])
        plt.xlabel(feature)
        plt.ylabel('quality')
        plt.show()
```

```python
## Now, We will see outliers
for feature in data.columns :
    dataset = data.copy()
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

```python
data.isnull().sum()
```

```python
## Feature Scaling
from sklearn.preprocessing import MinMaxScaler
feature_data = [feature for feature in data.columns if feature!='quality']
```

```python
## Separate target values 
y=data['quality']
x=data.drop('quality',axis=1)
```

```python
scaler =MinMaxScaler()
```

```python
scaler.fit(x)
```

```python
scaler.transform(x)
```

```python
data = pd.concat([data['quality'].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(x), columns=feature_data)],
                    axis=1)
```
